### 1.1 Introduction to Matplotlib

Matplotlib is a widely-used Python library for creating visualizations and plots. It was initiated in 2003 by John Hunter, an American neurobiologist, with the goal of providing a Python alternative to MATLAB software's plotting capabilities. 

Matplotlib has gained popularity for its ability to create a variety of 2D plots and charts, making it a favored choice for data science and machine learning visualization tasks. The library was designed to mimic MATLAB's command-style interface, making it familiar to users transitioning from MATLAB. Matplotlib's features and global styling options simplify the process of generating graphs and figures, making it a powerful, free, and user-friendly option for data visualization compared to MATLAB.

If we dive deeper into understanding the structure of a Matplotlib Figure object, we can see these parts
* Spines: These are lines that connect the tick marks on the axis.
* Title: It's like the name of the whole Figure.
* Legend: This tells us about what's shown in the plot.
* Grid: These are lines that extend the tick marks both up-down and left-right.
* X/Y axis label: Words that explain the X and Y axes below the lines.
* Minor tick: These are tiny marks between the big tick marks.
* Minor tick label: Little words that go with the minor tick marks.
* Major tick: These are the big value marks on the lines.
* Major tick label: Words that go with the major tick marks.
* Line: This is a type of drawing that connects data points with a line.
* Markers: These are like dots or symbols that show where each data point

### 1.3 Basic Plots in Matplotlib

In this part, we'll explore various basic plots. We'll cover bar charts, pie charts, stacked bar and area charts, histograms, box plots, violin plots, scatter plots, bubble plots, and line plots for time series data.

````python
# Import libraries

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
````

#### 1.3.1 Bar Chart

To make a vertical bar plot, you can use the `plt.bar(x, height, [width])` function. If you want horizontal bars, use `plt.barh()`.

**Key aspects:**

* x: Sets the x coordinates of the bars.
* height: Determines the bar heights.
* width (optional): Establishes the width of all bars; the default is 0.8.

````python
regions = ['A', 'B', 'C', 'D']
x = np.arange(len(regions))
bar_width = 0.4

plt.bar(x - bar_width / 2, [20, 25, 40, 10], width=bar_width, label='Wheat')
plt.bar(x + bar_width / 2, [30, 15, 30, 20], width=bar_width, label='Corn')

plt.xticks(x, regions)
plt.xlabel('Regions')
plt.ylabel('Yield')
plt.title('Crop Yields in Different Regions')
plt.legend()

plt.show()
````

#### 1.3.2 Pie Chart

The `plt.pie(x, [explode], [labels], [autopct])` function is used to generate a pie chart.

**Key aspects:**

* x: Defines the sizes of the slices in the pie.
* explode (optional): Sets the offset for each slice's radius. The explode array should match the length of the x-array.
* labels (optional): Assign labels to each slice.
* autopct (optional): Displays percentages within the slices using a specific format, like '%1.1f%%'.

In the following code example, we'll employ a pandas data frame to create a pie chart for visualizing water usage. 

````python
# Prepare dataset
data = {
    'Usage': ['Leak', 'Clothes Washer', 'Faucet', 'Shower', 'Toilet', 'Other'],
    'Percentage': [12, 17, 19, 20, 24, 8]
}

df = pd.DataFrame(data)

# Create figure
plt.figure(figsize=(4, 4), dpi=75)
# Create pie plot
plt.pie('Percentage', explode=(0, 0, 0.1, 0, 0, 0), labels='Usage', data=df, autopct='%.0f%%')
# Add title
plt.title('Water usage')
# Show plot
plt.show()
````

#### 1.3.3 Stack Plot and Stem Plot

**Stacked Bar Chart**

To make a stacked bar chart, you use the same `plt.bar` function as for regular bar charts. But here, for each stacked bar, you need to call the `plt.bar` function again and set the bottom parameter, starting from the second stacked bar.

````python
categories = ['Category A', 'Category B', 'Category C']
values_1 = [15, 30, 25]
values_2 = [10, 20, 30]

x = np.arange(len(categories))

plt.figure(figsize=(6, 4))
plt.bar(x, values_1, label='Value 1')
plt.bar(x, values_2, bottom=values_1, label='Value 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.xticks(x, categories)
plt.legend()
plt.show()
````

**Stacked Area Chart**

The `plt.stackplot(x, y)` function generates a stacked area plot.

**Key aspects:**

* x: Sets the x-values for the data series.
* y: Sets the y-values for the data series. For multiple series, use plt.stackplot(x, y1, y2, y3, …) with y1, y2, y3, and so on representing each data set.
* labels (optional): Assign labels to data series, using a list or tuple.

This part of the example is also adapted from the article [Data Visualization using Matplotlib](https://medium.com/towards-data-science/data-visualization-using-matplotlib-16f1aae5ce70). Download your dataset for the code example from [here](https://github.com/BadreeshShetty/Data-Visualization-using-Matplotlib/blob/master/datasets/national_parks.csv)

````python
np_data= pd.read_csv('data/national_parks.csv')

np_data.head()
````

````python
x = np_data['Year']
y = np.vstack([np_data['Badlands'], 
               np_data['GrandCanyon'], 
               np_data['BryceCanyon']])

# Labels for each stack
labels = ['Badlands', 
          'GrandCanyon', 
          'BryceCanyon']

# Colors for each stack
colors = ['sandybrown', 
          'tomato', 
          'skyblue']

#  Similar to pandas df.plot.area()
plt.stackplot(x, y, 
              labels=labels,
              colors=colors, 
              edgecolor='black')

# Plots legend to the upperleft of Figure
plt.legend(loc=2)

plt.show()
````

**Percentage Stacked Area Chart**

Stack plots based on percentage in distribution.

````python
plt.figure(figsize=(6,4))

data_perc = np_data.divide(np_data.sum(axis=1), axis=0)

plt.stackplot(x,
              data_perc["Badlands"],data_perc["GrandCanyon"],data_perc["BryceCanyon"],
              edgecolor='black',
              colors=colors,
              labels=labels)

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
````

**Stem Plots**

Similar to stack plots but stem plots help us view negative numbers, the `plt.stem()` function generates a stem plot. Use the `diff()` function to find the difference between consecutive data points.

````python
np_stem= np_data.copy()

np_stem[['Badlands', 
         'GrandCanyon',
         'BryceCanyon']] = np_data[['Badlands', 
                                    'GrandCanyon',
                                    'BryceCanyon']].diff()

np_stem.head()
````

````python
plt.figure(figsize=(15,11))

plt.suptitle('Change in Number of Visitors', y=0.94)

plt.subplot(311)
plt.stem(np_stem['Year'],
         np_stem['Badlands'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('GrandCanyon')

plt.subplot(312)
plt.stem(np_stem['Year'],
         np_stem['GrandCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('Badlands')

plt.subplot(313)
plt.stem(np_stem['Year'],
         np_stem['BryceCanyon'],
         markerfmt = 'b_',
         linefmt = 'r--',
         basefmt = 'g:')
plt.title('BryceCanyon')

plt.show()
````

#### 1.3.4 Histogram

A histogram shows how a single number changes. Each bar tells us how often numbers fall into a specific range. Use `plt.hist(x)` to build a histogram.

**Key aspects:**

* x: Put your data in here.
* bins (optional): Decide how many bins you want using a number, or use a list to specify bin edges.
* range (optional): Choose the lowest and highest values for the bins by using a tuple.
* density (optional): If you set this to true, the histogram will show probabilities.

````python
np_data=pd.read_csv('data/national_parks.csv')

plt.hist(np_data['GrandCanyon'], 
         facecolor='peru', 
         edgecolor='blue', 
         bins=10)
plt.show()
````

#### 1.3.5 Boxplot and Violin Plot

**Boxplot**

Boxplot gives a nice summary of the data. It helps in understanding our distribution better. To make a box plot, use `plt.boxplot(x)` function.

#### 1.3.6 Scatter Plot and Bubble Plot

**Scatter Plot**

Scatter plots reveal data points for two numeric variables, plotting one variable on each axis. To craft a scatter plot of y versus x, use `plt.scatter(x, y)`, and you can even change marker size and color if needed.

**Key aspects:**

* x, y: Show where your data points are located.
* s (optional): Decide marker size in points squared.
* c (optional): Pick marker color. If you give a sequence of numbers, it will be matched to colors from the color map.

In the following code example, you will download a [dataset](https://github.com/PacktWorkshops/The-Data-Visualization-Workshop/blob/master/Datasets/anage_data.csv) with details about different animals. Show how attributes like Maximum Longevity in years and Body Mass in grams are connected using visualizations.

````python
# Load dataset
data = pd.read_csv('data/anage_data.csv')

# Preprocessing
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass])]
# Sort according to class
amphibia = data[data['Class'] == 'Amphibia']
aves = data[data['Class'] == 'Aves']
mammalia = data[data['Class'] == 'Mammalia']
reptilia = data[data['Class'] == 'Reptilia']
````

````python
# Create figure
plt.figure(figsize=(10, 6), dpi=300)

# Create scatter plot
plt.scatter(amphibia[mass], amphibia[longevity], label='Amphibia')
plt.scatter(aves[mass], aves[longevity], label='Aves')
plt.scatter(mammalia[mass], mammalia[longevity], label='Mammalia')
plt.scatter(reptilia[mass], reptilia[longevity], label='Reptilia')
# Add legend
plt.legend()
# Log scale
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
# Add labels
plt.xlabel('Body mass in grams')
plt.ylabel('Maximum longevity in years')

# Show plot
plt.show()
````

**Bubble Plot**

The `plt.scatter` function is used to create a bubble plot. To visualize a third or fourth variable, the parameters s (scale) and c (color) can be used.

````python
# Generate random data
np.random.seed(42)
n = 20
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(30, 150, n)  # Slightly larger bubble sizes
colors = np.random.rand(n)  # Random colors

# Create bubble plot
plt.figure(figsize=(7, 5))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7)

# Add labels
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Bubble Plot Example')

# Show plot
plt.show()
````

#### 1.3.7 Line plot

A line plot, particularly in the context of time series data, connects data points using straight lines.  This is helpful for understanding patterns over time and uncovering relationships between the data points based on these trends. When the line moves upwards, it signifies a positive relationship, while a downward movement indicates a negative relationship. People often use time series plots to predict future trends and monitoring models.

In the following code example, we will create a line plot to show stock trends. Download your dataset [here](https://github.com/BadreeshShetty/Data-Visualization-using-Matplotlib/blob/master/datasets/stocks.csv).

## 2. Seaborn  <img src='images/05/seaborn_logo.svg' align='right' width=50>

### 2.1 Introduction to Seaborn

Unlike Matplotlib, Seaborn isn't a standalone Python tool. It's built on top of Matplotlib and offers a simpler way to create good-looking statistical graphs. One cool thing about Seaborn is that it can work together with DataFrames from the pandas library.

With Seaborn, the goal is to put visualization at the heart of exploring and understanding data. Inside, Seaborn works with DataFrames and arrays that hold all the data. This lets it make sense of the data and statistics to create useful graphs. You can also use Seaborn just to change how Matplotlib graphs look.

The most important features of Seaborn are:

* Instant attractive graphs with different styles
* Ready-to-use color choices that help show patterns in the data
* A way to work directly with datasets
* A simpler way to make complex graphs while still keeping things easy to use.

### 2.2 Advantages of Seaborn

Here are several advantages of Seaborn:

1. **Easy Data Handling:** Unlike Matplotlib, Seaborn is tailored for DataFrames, making plotting simpler. It manages data manipulation internally, creating informative plots effortlessly.

2. **Simple Variable Usage:** Seaborn uses Matplotlib for plotting, but it's easier. You only need variable names, no complex translation like Matplotlib.

3. **Better Looks by Default:** Seaborn's default style is visually pleasing, better than Matplotlib's basic look. It has ready-to-use themes and user-friendly ways to tweak appearance.

4. **Clear Parameter Management:** Seaborn sorts Matplotlib settings into two groups: style and scaling. It adds to Matplotlib without complexity, offering pre-made plots like Facet and Regression.

5. **Quick Regression Plots:** Seaborn excels in making Regression plots compared to Matplotlib. It's simpler and shows visually appealing data relationships.

6. **Familiarity with Matplotlib Helps:** If you know Matplotlib, Seaborn's concepts are familiar. This makes it easy to harness Seaborn's benefits.

### 2.3 Color Palettes

Picking the right colors for your figures is really important. It can make patterns in the data clear when done well, or make them hard to see if not done right. Even people who know their stuff sometimes think using colors is easy. They just pick a set of colors, like shades of gray or a rainbow, choose where they begin and end, and use them. But it's not as simple as that. Sometimes this doesn't work out, and the graphs don't show the data as well as they could. The main idea in choosing colors is to make sure that important data points in big sets of data can be seen easily.


Picking colors based on numbers is a key aspect. It helps us add colors to things using different numbers. The rainbow color map is common in science visuals. If you're interested, you can check out a detailed research paper called "[Diverging Color Maps for Scientific Visualization](https://www.kennethmoreland.com/color-maps/)" by Kenneth Moreland.

Let's now shift our attention to what Seaborn offers. But before we do that, remember that Seaborn works on top of Matplotlib, so any color supported by Matplotlib will also work with Seaborn. Before diving into Seaborn's features, let's understand how colors work in Matplotlib. You can use various ways to define colors:

* Numbers between 0 and 1 in an RGB or RGBA format (like (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
* A code like ‘#0F0F0F’ for RGB or ‘#0F0F0F0F’ for RGBA
* A grayscale level between 0 and 1, like ‘0.5’
* Letters like ‘b’, ‘g’, ‘r’ for blue, green, and red
* Names like ‘xkcd:sky blue’ from the xkcd color survey
* Abbreviations like ‘C0’ or ‘tab:blue’ for pre-defined colors

#### 2.4.1 Bar Plot

Making bar plots with subgroups used to be a bit of a hassle, but Seaborn simplifies this process significantly. You can also use them in Seaborn to show estimates of the middle point using bar height, and any uncertainty is shown with error bars at the top of each bar

For the code example, we'll utilize a dataset that contains information about salaries across various age groups, English language proficiency, education levels, and gender. You can download the dataset [here](https://github.com/PacktWorkshops/The-Data-Visualization-Workshop/blob/master/Datasets/salary.csv).

````python
import pandas as pd
import seaborn as sns

data = pd.read_csv("data/salary.csv")
data = data.drop(data.columns[0], axis=1)
data.head()
````

````python
# Create a bar plot
sns.set(style="darkgrid")
sns.barplot(x="Education", y="Salary", hue="District", data=data)
````

#### 2.4.2 Kernel Density Estimation

Visualizing the distribution of variables within a dataset can be quite insightful. Seaborn provides useful tools to explore both single and paired distributions. A simple approach to viewing a single-variable distribution in Seaborn is by using the `histplot()` function. This creates a histogram and overlays a kernel density estimate (KDE), as demonstrated in the following example.

We will still use the salary dataset.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()  # Set Seaborn style
sns.histplot(data['Age'], kde=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
````

If you only want to see the KDE (Kernel Density Estimate) plot, Seaborn offers the `kdeplot(`) function. The KDE plot displays a shaded region under the curve.:

````python
sns.kdeplot(data.loc[:, 'Age'], shade=True)
plt.xlabel('Age')
plt.ylabel('Density')
````

#### 2.4.3 Plotting Bivariate Distribution

To explore bivariate distributions visually, we'll discuss three distinct plots. The initial two plots utilize the `jointplot()` function, generating a multi-panel figure that illustrates the joint correlation between two variables alongside their respective marginal distributions.

Considering the extensive size of the salary dataset, we'll establish a smaller subset by selecting the initial 10,000 rows.

````python
import matplotlib.ticker as ticker

subdata = data[0:999]

sns.set(style="white")
joint = sns.jointplot(x="Salary", y="Age", data=subdata)

# Rotating x-axis labels for better readability
joint.ax_joint.xaxis.set_tick_params(rotation=45)

plt.show()
````

You can also use the KDE technique to show bivariate distributions. This presents the combined distribution using a contour plot, as you can see in the code example below.

````python
joint = sns.jointplot(data=subdata, x="Salary", y="Age", kind='kde', fill=True, xlim=(0,250000), ylim=(0, 100))

# Rotating x-axis labels for better readability
joint.ax_joint.xaxis.set_tick_params(rotation=45)

plt.show()
````

The center of the diagram displays the joint distribution as a contour plot. The color's darkness indicates higher density. Additionally, the top and right sides of the diagram illustrate the marginal distributions.

#### 2.4.4 Violin Plot

A different way to show statistical information is through violin plots. These plots combine box plots with the kernel density estimation we discussed earlier. This approach gives a more detailed view of how the variable is distributed. Plus, the violin plot also includes the quartile and whisker values from the box plot.

#### 2.4.6 Regression Plots

Regression means figuring out how one variable (often shown on the Y-axis) is linked to another variable (usually shown on the X-axis). When we have data, we can choose these variables and use regression techniques to grasp their connection. For now, we'll stick to talking about linear regression, although Seaborn has more options for other types of regression.

To see the relationships learned by linear regression, Seaborn provides the `regplot()` function. Here's a simple example.

````python
import numpy as np
import seaborn as sns

x = np.arange(100)
# Creating a normal distribution with a mean of 0 and a standard deviation of 5
y = x + np.random.normal(0, 10, size=100) 

# Using Seaborn's regplot() function to visualize the linear relationship
sns.set(style="darkgrid") 
sns.regplot(x=x, y=y)

# Display the plot
plt.show()
````

#### 2.4.7 Multi-Plot

Frequently, it's beneficial to showcase multiple plots side by side. Matplotlib introduces the concept of `subplots()`, which are essentially multiple Axes residing within a single Figure. These subplot arrangements can include grids of plots, nested plots, and more. In Seaborn, the `factorplot()` function sets up a grid of multiple plots to visualize conditional relationships. 

Let us see some code examples.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("data/age_salary_hours.csv")
degrees = set(data['Education'])
ordered_degrees = sorted(list(degrees))

# Set color palette to colorblind
sns.set_palette('colorblind')

# Create subplot with two rows
fig, ax = plt.subplots(2, 1, figsize=(8, 8))

sns.violinplot(x='Education', y='Annual Salary', data=data, cut=0, order=ordered_degrees, ax=ax[0])
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=10)
ax[0].set_title('Impact of Education on Annual Salary', fontsize=12)  # Change title font size

sns.violinplot(x='Education', y='Weekly hours', data=data, cut=0, order=ordered_degrees, ax=ax[1])
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=10)
ax[1].set_title('Impact of Education on Weekly Working Hours', fontsize=12)  # Change title font size

plt.tight_layout()

# Show figure
plt.show()
````

````python
data = pd.read_csv("data/salary.csv")
data = data.drop(data.columns[0], axis=1)

# Create a multi-plot grid using catplot
g = sns.catplot(x='Education', y='Salary', hue='Gender', row='District', col='English',data=data, kind='violin')

# Add title and adjust layout
g.fig.suptitle('Salary by Education and Gender')
plt.subplots_adjust(top=0.85)

# Show the plot
plt.show()
````
